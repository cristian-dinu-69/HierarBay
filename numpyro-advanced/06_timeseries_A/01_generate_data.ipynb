## Generate Data

Smooth Trend Model を使って、以降の解析に利用する時系列データを生成する

## Import Packages

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (8, 4)
```

## Set Parameters

```python
num_steps = 100
step_size = 0.1
```

## Normal Random Numbers

```python
np.random.seed(0)
```

```python
z = np.random.randn(100) * step_size

plt.plot(z)
plt.title('Normal Random Numbers');
```

```python
zz = np.random.randn(100, 10) * step_size

plt.plot(zz)
plt.title('Normal Random Numbers');
```

## Random Walk

```python
b = np.cumsum(z)

plt.plot(b)
plt.title('Random Walk');
```

```python
bb = np.cumsum(zz, axis=0)

plt.plot(bb)
plt.title('Random Walk');
```

## Random Trend

```python
u = np.cumsum(b)

plt.plot(u)
plt.title('Random Trend');
```

```python
uu = np.cumsum(bb, axis=0)

plt.plot(uu)
plt.title('Random Trend');
```

## Generate Data

```python
sigma = 5

c = 10
e = np.random.randn(num_steps)

u = u + c
y = u + e * sigma

plt.plot(u)
plt.plot(y, '.')
plt.title('Random Trend');
```

## Save Data

```python
t = np.arange(len(y))

data = pd.DataFrame({'t':t, 'u':u, 'y':y})

data.to_csv('data.csv', index=False)
```
