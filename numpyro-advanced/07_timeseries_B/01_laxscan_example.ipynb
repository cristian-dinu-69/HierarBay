## Scan Function


## Import Packages

```python
import jax
import jax.numpy as jnp

from jax.lax import scan
```

## Define Data

```python
vec_in = jnp.arange(10)
vec_in
```

## Example-1
入力をそのまま２倍して出力する例（carry は使わない）

```python
def transition_fn(carry, x_in):
    
    x_out = 2 * x_in
    
    return carry, x_out
```

```python
carry_init = None

carry_ends, vec_out = scan(transition_fn, carry_init, vec_in)
```

```python
vec_out
```

```python
carry_ends
```

## Example-2
carry に入力値を加算する例（出力は carry の値を出力）

```python
def transition_fn(carry, x_in):
    
    carry = carry + x_in
    
    x_out = carry
    
    return carry, x_out
```

```python
carry_init = 0

carry_ends, vec_out = scan(transition_fn, carry_init, vec_in)
```

```python
vec_out
```

```python
carry_ends
```

## Example-3
carry に２つの状態を持たせた例（出力は carry の値を出力）

```python
def transition_fn(carry, x_in):
    
    state0, state1 = carry
    
    state0 = state0 + x_in
    state1 = state1 + 2 * x_in
    
    carry = (state0, state1)
    x_out = carry
    
    return carry, x_out
```

```python
carry_init = (0, 0)

carry_ends, vec_out = scan(transition_fn, carry_init, vec_in)
```

```python
vec_out
```

```python
carry_ends
```
